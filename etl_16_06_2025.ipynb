```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
print(df.columns.tolist())
```

```python
# Remover colunas sem nome
df = df.loc[:, df.columns.notna()]
```

```python
# Remover colunas com nome vazio ou None
df = df.loc[:, df.columns.notnull() & (df.columns != '')]
```

```python
print(df.columns.tolist())
```

```python
query = """
SELECT *
FROM df
WHERE DESCONTO > 1000
ORDER BY DESCONTO DESC
"""

resultado = pysqldf(query)
resultado.head()
```

```python
query = """
SELECT *
FROM df
ORDER BY "%" DESC
LIMIT 10
"""

resultado = pysqldf(query)
resultado.head()
```

```python
query = """
SELECT SITUAÇÃO, COUNT(*) AS TOTAL
FROM df
GROUP BY SITUAÇÃO
ORDER BY TOTAL DESC
"""
resultado = pysqldf(query)
resultado.head()
```

```python
query = """
SELECT *
FROM df
WHERE "%" BETWEEN 70 AND 80
ORDER BY "%" DESC
"""
resultado = pysqldf(query)
resultado.head()

```

```python
query = """
SELECT *
FROM df
ORDER BY "DESCONTO" DESC
LIMIT 10
"""
resultado = pysqldf(query)
resultado.head()
```

```python
query = """
SELECT CONSULTOR, COUNT(*) AS TOTAL_CONTRATOS
FROM df
GROUP BY CONSULTOR
ORDER BY TOTAL_CONTRATOS DESC
"""
resultado = pysqldf(query)
resultado.head()
```

```python
query = """
SELECT BANCO,
       AVG(CAST(REPLACE(REPLACE(REPLACE("%", '%', ''), ',', '.'), ' ', '') AS FLOAT)) AS MEDIA_PORCENTAGEM
FROM df
GROUP BY BANCO
ORDER BY MEDIA_PORCENTAGEM DESC
"""
resultado = pysqldf(query)
resultado.head()
```

```python
query = """
SELECT UF, COUNT(*) AS TOTAL
FROM df
GROUP BY UF
ORDER BY TOTAL DESC
"""
resultado = pysqldf(query)
resultado.head()
```
